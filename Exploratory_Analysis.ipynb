```python
import pandas as pd
import numpy as np
from scipy import stats
df = pd.read_csv("train.csv")
print(df)
```

```python
numerical_cols = df.select_dtypes(include=[np.number])
Q1 = numerical_cols.quantile(0.25)
Q3 = numerical_cols.quantile(0.75)

# Calculate IQR
IQR = Q3 - Q1

# Define outliers as points outside the IQR range
outliers_iqr = ((numerical_cols < (Q1 - 1.5 * IQR)) | (numerical_cols > (Q3 + 1.5 * IQR)))

# Show rows with outliers
outlier_rows_iqr = df[outliers_iqr.any(axis=1)]
print(outlier_rows_iqr)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.boxplot(data=numerical_cols)
plt.xticks(rotation = 90)
plt.show()

plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['Posts_Per_Day'])
plt.xlabel('Age')
plt.ylabel('Posts_Per_Day')
plt.show()
```

```python
from google.colab import files
uploaded = files.upload()

import pandas as pd

df = pd.read_csv('train.csv')
print(df)

numerical_cols = df.select_dtypes(include=[np.number])

column = numerical_cols.columns[0]
data = numerical_cols[column]

minimum = data.min()
Q1 = data.quantile(0.25)
median = data.median()
Q3 = data.quantile(0.75)
maximum = data.max()

five_num_summary = {
    'min': minimum,
    'Q1': Q1,
    'median': median,
    'Q3': Q3,
    'max': maximum
}

print(five_num_summary)

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.boxplot(data=data)

plt.text(0, minimum, f'min: {minimum:.2f}', horizontalalignment='right')
plt.text(0, Q1, f'Q1: {Q1:.2f}', horizontalalignment='right')
plt.text(0, median, f'median: {median:.2f}', horizontalalignment='right')
plt.text(0, Q3, f'Q3: {Q3:.2f}', horizontalalignment='right')
plt.text(0, maximum, f'max: {maximum:.2f}', horizontalalignment='right')

plt.title(f'Box Plot of {column}')
plt.show()
```

```python
import pandas as pd

df = pd.read_csv("train.csv")

print(df.head())

print(df.info())

print(df.describe())


```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd



sns.scatterplot(x='Posts_Per_Day', y='Messages_Sent_Per_Day', data=df)
plt.title('Scatter Plot')
plt.show()

plt.plot(df['Posts_Per_Day'], df['Comments_Received_Per_Day'])
plt.title('Line Plot')
plt.show()

plt.hist(df['Posts_Per_Day'], bins=10, color='blue', alpha=0.7)
plt.title('Histogram')
plt.show()


sns.boxplot(x='Age', y='Posts_Per_Day', data=df)
plt.title('Box Plot')
plt.show()


sns.barplot(x='Age', y='Posts_Per_Day', data=df)
plt.title('Bar Plot')
plt.show()



plt.pie(df['Age'].value_counts(), labels=df['Age'].unique(), autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart')
plt.show()
```
